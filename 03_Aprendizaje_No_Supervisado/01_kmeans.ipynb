# 🔵 K-Means Clustering

**K-Means** agrupa los datos en K clústeres asignando cada punto al centroide más cercano e iterativamente actualizando los centroides.

**Algoritmo:**
1. Inicializar K centroides aleatoriamente.
2. Asignar cada punto al centroide más cercano.
3. Recalcular los centroides como la media de los puntos asignados.
4. Repetir hasta convergencia.

**Función objetivo (inercia):**
$$J = \sum_{i=1}^{K} \sum_{\mathbf{x} \in C_i} \|\mathbf{x} - \boldsymbol{\mu}_i\|^2$$

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score

sns.set_theme(style='whitegrid')
np.random.seed(42)
print("Librerías importadas ✅")
```

## 1. Dataset Sintético

```python
# Generar 3 grupos bien separados
X, y_true = make_blobs(n_samples=300, centers=3, cluster_std=0.8, random_state=42)

plt.figure(figsize=(7, 5))
plt.scatter(X[:, 0], X[:, 1], c='gray', alpha=0.5, s=30)
plt.title('Datos sin etiquetar')
plt.show()
```

## 2. Elegir K — Método del Codo

```python
inercias = []
siluetas = []
ks = range(2, 9)

for k in ks:
    km = KMeans(n_clusters=k, n_init=10, random_state=42)
    labels = km.fit_predict(X)
    inercias.append(km.inertia_)
    siluetas.append(silhouette_score(X, labels))

fig, axes = plt.subplots(1, 2, figsize=(13, 4))

axes[0].plot(ks, inercias, 'bo-')
axes[0].set(xlabel='Número de clústeres (K)', ylabel='Inercia', title='Método del Codo')

axes[1].plot(ks, siluetas, 'rs-')
axes[1].set(xlabel='Número de clústeres (K)', ylabel='Silhouette Score', title='Coeficiente de Silueta')

plt.tight_layout()
plt.show()

print(f"Mejor K por silueta: {ks[np.argmax(siluetas)]}")
```

## 3. Aplicar K-Means con K=3

```python
km = KMeans(n_clusters=3, n_init=10, random_state=42)
labels = km.fit_predict(X)
centroides = km.cluster_centers_

plt.figure(figsize=(8, 5))
scatter = plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='Set1', alpha=0.6, s=30)
plt.scatter(centroides[:, 0], centroides[:, 1], c='black', s=200, marker='X', zorder=5, label='Centroides')
plt.title('K-Means Clustering (K=3)')
plt.legend()
plt.colorbar(scatter, label='Clúster')
plt.show()

print(f"Inercia final: {km.inertia_:.2f}")
print(f"Silueta:       {silhouette_score(X, labels):.4f}")
```

## 4. Limitaciones de K-Means

- Asume clústeres **esféricos** y de tamaño similar.
- Sensible a **outliers** y a la inicialización.
- Hay que especificar **K** de antemano.
- No funciona bien con clústeres de formas no convexas → usar DBSCAN en esos casos.
