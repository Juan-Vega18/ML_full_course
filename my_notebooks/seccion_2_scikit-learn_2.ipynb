#Asegurarnos de que todo es numerico

Es muy importante que toda la data sea numérica, esto incluso en datasets los cuales contienen variables no-numericas que podriamos querer incluir en el modelo. 

Por ejemplo, si estamos trabajando con un dataset de car sales, como podriamos convertir variables como Make and Colour into numbers. 
Veamos. 


````python
url = 'https://raw.githubusercontent.com/Juan-Vega18/ML_full_course/refs/heads/main/my_notebooks/data/car-sales-extended.csv'
car_sales = pd.read_csv(url)
car_sales
````

````python
#Revisamos los tipos de variable: 
print(car_sales.dtypes)
print(car_sales.info())
print(car_sales['Make'])
print(car_sales['Colour'])
````

Podemos darnos cuenta que tenemos dos variables categóricas, tipo de variable object, ya previamente nos dimos cuenta que son las marcas y los colores de los carros. Si queremos usar esas columnas en nuestro modelo necesitamos convertirlas en numericas. 

````python
X = car_sales.drop('Price', axis=1)
y = car_sales['Price']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

#Ahora construimos el modelo 

from sklearn.ensemble import RandomForestRegressor
 
model = RandomForestRegressor()
model.fit(X_train, y_train)
model.score(X_test, y_test)
````

Aquí nos damos cuenta que el modelo no logró convertir la string a un float, entonces, tenemos que convertir las variables en numeros primero. El proceso de convertir las variables categoricas en numericas es usualmente llamado: encoding. 

Scikit-Learn tiene una guia excelente para esto: 
https://scikit-learn.org/stable/modules/preprocessing.html#encoding-categorical-features

Pero vamos a las maneras mas directas para convertir las variables categoricas en numericas, one-hot encoding. 

En machine learning, one-hot enconding le da un valor de 1 al valor objetivo y 0 a otros valores. Por ejemplo, tenemos 3 marcas de carros: Honda, Toyota, BMW: 

Nuestros samples son: 
1. HONDA 2. BMW 3. BMW 4. TOYOTA 5. 

| Sample | Honda | Toyota | BMW |
| ----- | ----- | ----- | ----- |
| 1 | 1 | 0 | 0 |
| 2 | 0 | 0 | 1 |
| 3 | 0 | 0 | 1 |
| 4 | 0 | 1 | 0 |
| 5 | 0 | 1 | 0 |


Entonces, da 1 para cada valor objetivo, pero 0 para cada otro valor. 
Podemos seguir los siguientes pasos para hacer el one-hot encode en el dataset. 

1. Import [`sklearn.preprocessing.OneHotEconder`] to one-hot encode our features and [`sklearn.compose.ColumnTransformer`] to target the specific columns of our DataFrame to transform. 

2. Definir las variables categoricas que nos gustaria transformar
3. Creamos una instancia de OneHotEncoder
4. Creamos una instancia de ColumnTransformer y le ponemos las transformaciones que nos gustaria hacer. 
5. Llenamos the instance of the `ColumnTransformer` para nuestra data y la transformamos con `fit transform(X)` method. 




````python
#1. Importamos OneHotEncoder and ColumnTransformer 

from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

#2. Define the categorical features to transform 
categorical_features = ['Make', 'Colour', 'Doors']

#3. Crea an instance of OneHotEncoder
one_hot = OneHotEncoder() #--> Iniciamos la instancia

#4. Creamos la instancia del ColumnTransformer
transformer = ColumnTransformer([("one_hot", # name
                                  one_hot, # transformer
                                  categorical_features)], # columnas para transformar
                                remainder='passthrough') # ¿Que haras con el resto de las columnas? 
                                # 'passtrough' = leave unchanged 

#5. Convertimos las variables categoricas en numericas.
# esto devolverá una array-like sparse matrix, not a DataFrame

transformed_X = transformer.fit_transform(X)
transformed_X
````

Bueno, aqui vale muuuuuchiiiisimo la pena explicar que está pasando porque no es trivial. Algo que no se suele explicar mucho es el hecho de convertir variables categoricas en numericas para poder ejecutar los modelos. En el conocimiento basico que uno como estudiante suele tener uno dice: facil, uso `df['Make'] = df['Make'].astype(int)` y sale, pero, la pregunta sincera es... que es realmente lo que está pasando, el analisis no es tan simple como parece. 

Lo que nosotros tenemos es una funcion de n variables (features) y cada variable tiene su propio espacio en el que vive. En el caso de 'Make' que son las marcas, tenemos 4 marcas, entonces, si yo quiero mapear las marcas a numeros con algun diccionario, introduzco un error fuerte conceptual y es: ¿Por qué habria un orden? ¿Como escogerlo? y aparte metemos una jerarquía inexistente. 

Entonces, en terminos simples, lo que hace el one-hot encoding es expresas cada sub_feature como un vector de, en este caso, R^4. Luego de eso, simplemente si teniamos la feature 'Make' ahora tenemos 'Make_Toyota', 'Make_Nissan' etc. entonces, crece el dataset en columnas, pero no en complejidad, todas son binarias. 
Asi mismo con las puertas `Doors` del carro, etc. 

````python
print('El dataset transformado es: ')
print(transformed_X[0]) #Vemos la primera fila (usamos corchetes solo)
print('El dataset original era: ')
print(X.iloc[0]) #Vemos la primera fila (usamos iloc)
print(f'El shape del dataset transformado es: {transformed_X.shape}') #Vemos tamaño luego de las transformaciones
print(f'El shape del dataset original era: {X.shape}') # Tamaño original 
````

Como podemos notar, pasamos de 4 columnas a 13, preservamos la informacion del modelo sin introducir un orden jerarquico irreal... Ahora, obviamente si tendremos casos en los que existe variable categorica ordinal (como el estrato). Y eso es precisamente lo que sucede cuando hacemos un: 
`X['Make'] = X['Make'].astype(int)`

Pandas hace un label encoding jerarquico tomando en cuenta el orden alfabetico...Nosotros mismo podriamos resolver eso usando un mapa: 

`[size_mapping = {
    "Small": 0,
    "Medium": 1,
    "Large": 2
}]`

`[X["size"] = X["size"].map(size_mapping)]`

#Que pasa si tenemos valores faltantes? 

Vacíos en la data significan vacios en los patrones que el modelo de machine learning puede aprender. 
Muchos de os modelos de machine learning no funcionan bien o producen errores cuando se entrenan con datasets con valores faltantes.  

Un valor faltante puede aparecer como blank (vacío), tambien como `NaN`, o algo similar. Hay dos opciones principalmente disponibles para manejar los valores faltantes. 

1. Llenarlos con algo dado o un valor calculado (imputation). Por ejemplo, podrias llenar los valores numericos faltantes en una columna con la media de los otros valores, la practica de calcular o figuring como llenar los valores faltantes en un dataset es llamada imputing. Scikit-Learn tiene grandes recursos para aprender sobre esto. 

2. Removerlos. Si una fila o un ejemplo tiene valores faltantes podriamos optar por removerlos de golpe del dataset. Sin embargo, esto puede significar varias cosas: 1. Tener menos datos para hacer el entrenamiento y 2. perder la oportunidad de entender porque estan vacíos esos datos. 

En conclusion, lidiar con los valores faltantes no es algo que tenga al 100% una unica solucion, depende mucho. del caso en el que estemos trabajando.

#Filling missing data and transforming categorical data with Scikit-Learn. 
Ya que lo hicimos con pandas... volvemos a la misma idea, Pandas es genial para datos pequeños y reparaciones "simples" pero, ahora necesitamos llamar a las herramientas mas robustas. 

`sklearn.impute.SimpleImputer()` el cual nos permite hacer algo similar en los que hicimos con pandas. 
`SimpleImputer()` nos permite transformar la data llenando los valores missing con una given `strategy` parameter. 
Vamos a reimportar el dataset y hacerlo. 

````python
car_sales_missing = pd.read_csv('https://raw.githubusercontent.com/Juan-Vega18/ML_full_course/refs/heads/main/my_notebooks/data/car-sales-extended-missing-data.csv')
print(car_sales_missing.head(10))
print('#'*80)
print('\n')
print(car_sales_missing.info(), car_sales_missing.shape)
car_sales_missing.isna().sum()
````

````python
#Para empezar, vamos a ahora si remover los NaN de Price (eso de quitarlos en la columna target es mas stantard)

car_sales_missing.dropna(subset=['Price'], inplace=True)
car_sales_missing.isna().sum()
````

````python
#Ahora dividimos la data en X e Y (como siempre)

X = car_sales_missing.drop('Price', axis=1)
y = car_sales_missing['Price']
#Dividimos en sets de entrenamiento 
np.random.seed(42)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

y.isna().sum()
````

Aqui lo quie hicimos fue separar la data en traing y test sets para hacer el filling de los valores faltantes por separado. Que de hecho, es la mejor practica dado que se supone que el test set es data que el modelo nunca ha visto. Para las variables categoricas en general está bien llenarlas con valores a lo largo del dataset (completo, no importa), pero para las numericas solo deberia llenarse los valores en el test set con valores que hayan sido calculados con el training set, no al revés. 

Vamos a ver algunas funciones del `SimpleImputer()` para llenar varios valores faltantes. 
1. For categorical columns (Make, Colour) `strategy='constant'`, `fill_value = 'missing'` (llenamos los samples vacios con un valor consistente valor de missing)

2. Para `Door` column, `strategy ='constant`, `fill_value=4` (igual a pandas pero ahora con Sklearn)

3. For numerical column (`Odometer KM`) `strategy = 'mean'`

Hay mas estrategias... en la documentacion de `SimpleImputer()` se pueden encontrar. 

````python
from sklearn.impute import SimpleImputer

#Create a categorical variable imputer
cat_imputer = SimpleImputer(strategy='constant', fill_value='missing')
#Create Door column imputer
door_imputer = SimpleImputer(strategy='constant', fill_value=4)
#Create Odometer (KM) column imputer 
num_imputer = SimpleImputer(strategy='mean')

#Creados los imputer, ahora, definimos las columnas donde queremos aplicarlos

categorical_features = ['Make', 'Colour']
door_feature = ['Doors']
numerical_feature = ['Odometer (KM)']
````

Ahora. ¿Como aplicamos las transformaciones a nuestras columnas? Podemos usar `sklearn.compose.ColumnTransformer` class de Scikit-Learn de forma similar a como lo hicimos antes para convertir la data en numerica. 

ColumnTransformer() toma una lista de entrada tuplas de forma (nombre_a_transformar, transformer_to_use, columns_to_transform) specifying which columns to transform and how transform them. por ejemplo 

```python
imputer = 
ColumnTransformer([
    'cat_imputer', 
    cat_imputer, 
    categorical_features])

```

En nuestro caso, las variables en la tupla seran: 

name_of_transform = "cat_imputer", transformer_to_use = cat_imputer, columns_to_transform (la lista que definimos arriba)

````python
from sklearn.compose import ColumnTransformer
#Crear la serie de transformacion para cada columna

imputer = ColumnTransformer([
    ("cat_imputer", cat_imputer, categorical_features),
    ("door_imputer", door_imputer, door_feature),
    ("num_imputer", num_imputer, numerical_feature)
])

#Columnas definidas. 
````

El siguiente paso es ajustar nuestra instancia de ColumnTransformer() `(imputer)`,  a la data de entrenamiento y transformar nuestra data de prueba

En otras palabras, queremos: 
1. Learn the imputation values from the training set.
2. Fill the missing values in the training set with the values learned in 1. 
3. Fill the missing values in the testing set with the values learned in 1. 

Es decir, para llenar valores en el test_set solo se pueden usar valores calculados o reemplazados hallados en el training_set. Noto que se hace mucho incapié en esto. 

Podemos hacer el paso 1 y 2 simultaneamente con `ColumnTransformer.fit_transform()` method, fit=find the values to fill. `transform = fill them`. 
Y podemos hacer el paso 3 con `ColumnTransformer.transform()` method (solo queremos transformar el test set, no aprender diferentes valores para rellenar)

````python
#Find values to fill and transform training data. 

filled_X_train = imputer.fit_transform(X_train)
filled_X_test = imputer.transform(X_test)
filled_X_train.shape

#Check filled X_train
#filled_X_test
````
