#QUE ES LA CROSS-VALIDATION
Es ua forma de asegurar que los resultados que se estan obteniendo son consistentes a lo largo de los intentos y pruebas de entrenamiento, usar multiples versiones de entrenamiento y sets de datos influye, en lugar de solo suerte por el orden de el set de datos de entrenamiento y de pruebas que se crearon. 
Hay dos elementos importantes: 
1. Probar diferentes hiperparámetros 
2. Todos los diferentes parametros deben ser cross-validated. (Es importante recordar que si tenemos series de tiempo no se deben mezclar) 

En este caso, nuestro modelo RandomForestClassifier(), vamos a empezar probando diferentes valores de n_stimators que basicamente es la cantidad de arbol en el Rando Forest. 
Por defecto tenemos n_estimators = 100, entonces, vamos a intentar valores entre 100 y 200 y veremos que sucede.. ¿Generalmente es mejor? 


```python
np.random.seed(42)
for i in range(100,200,10):
    print(f'Trying model with {i} estimators... ')
    model = RandomForestClassifier(n_estimators=i).fit(X_train, y_train)
    print(f'Model accuray on test set: {model.score(X_test, y_test)}')
    print("")
```

Notamos que cuando usamos diferente numero de estimadores obtenemos diferentes puntajes. En el caso de las pruebas de 10 en 10 encontramos que con 110, 120, 150, 170 encontramos que tenemos el mismo score. Por lo tanto, no es cuestión aumentar los estimadores arbitrariamente. 

Usaremos sklearn.model selection.cross val score para medir el resultado con 5 diferentes sets de entrenamiento y prueba. (train and test sets)

Podemos medir esto configurando: 
cross_val_score(X, y, cv=5)
Ddonde X el dataset completo, y es el full set de etiquetas (basicamente el dato predictivo) y cv es el numero de sets de entrenamiento y test splits cross_val_score creará automaticamente de nuestra data (en nuetro caso) 5 diferentes divisiones, esto es conocido como 5-fold cross-validation. 

```python
from sklearn.model_selection import cross_val_score

#With cross validation 
np.random.seed(42) #Elegimos una semilla para que el calculo sea rep1licable. 
for i in range(100,200,10):
    print(f'The model with {i} estimators...')
    model = RandomForestClassifier(n_estimators=i)

    scores = cross_val_score(model, X_train, y_train, cv=5)
    print(f'Scores: {scores}')
    print(f'Mean accuracy: {scores.mean()*100:.5f}%')
    print("")
```

¿Que modelo tiene la mejor cross-validation score? Este es usualmente un mejor indicador de la calidad del modelo que un unico puntaje de precisión. 
En lugar de hacer el set up y seguir los resultados de los experimentos manualmente, podemos hacer que Scikit-Lean haga la exploracion por nosotros. 

Scikit-Learn's sklearn.model selection.GridSearch CV es una forma de de checar diferentes hiperparametros y valores y evaluar automaticamente cual rinde mejor. 

```python
#Another way to do it with GridSearchCV...
np.random.seed(42)
from sklearn.model_selection import GridSearchCV

#Definimos los parametros para probar y los metemos en un diccionario
# estos pueden ser cualquiera de los hiperparametros del modelo 
param_grid = {'n_estimators': [i for i in range(100,200,10)]}

#Setup de grid search 

grid = GridSearchCV(estimator=RandomForestClassifier(), 
                    param_grid=param_grid,
                    cv=5,
                    verbose=1)

#Fit the grid search to the data
grid.fit(x,y) #Metemos la data completa

#Find the best parameters: 

print(f'The best parameter values are: {grid.best_params_}')
print(f'With a score of: {grid.best_score_*100:.2f}%')
```

Y aqui podemos fijarle al modelo el mejor estimador

```python
clf = grid.best_estimator_
clf
```

Me gustaria tomarme unos parrafos para explicar realmente que el cross-validation ya que creo que es muy importante tenerlo bien claro. Nosotros tenemos un set de data completo D. Primero se divide en 2 subconjuntos D_{train} (\approx80%) y D_{test} (\approx 20%). El conjunto D_{test} se reserva completamente y no se toca durante el entrenamiento ni la selección del modelo. 

Sobre D_{train} se aplica k-fold cross validation, entonces, que se hace:
1. Se divide D_{train} en k subconjuntos que llamaremos folds, todos del mismo tamaño 
2. Para cada iteracion segun el numero de k folds: 
2.1 Se entrena el modelo con k-1 folds 
2.2 Se evalua con con el fold restante. 
Esto singifica basicamente. tenemos FOLD1, FOLD2, FOLD3, FOLD4, FOLD5
Primer entrenamiento: TRAIN = FOLD1,FOLD3,FOLD4,FOLD5. TEST = FOLD2 y así hasta que se hayan bajarado todos. 
3. Se obtienen k metricas, como accuracy o score. 
4. Se calcula el promedo de las k metricas. 

Y este promedio ESTIMA el rendimiento esperado del modelo sobre datos no vistos. 
Y aquí hay algo clave y brutalmente importante. CROSS-VALIDATION NO GARANTIZA QUE EL TEST SEA REPRESENTATIVO. 

Simplemente nos sirve para reducir la incertidumbre en la estimación del modelo antes de evaluarlo en el test final. 

Y Ahora, ya que dominamos cross-validation (al menos de forma teórica) tenemos que pasar con esta idea base: 
'Si ya entendí que un modelo no debe evaluarse con una sola partición, entonces tampoco debería elegir sus hiperparámetros "a ojo" con una sola partición'. 
Y eso es lo que hace GridSearchCV

Estamos variando los n_estimators (es solo uno de los hiperparametros que tenemos disponibles) y luego comprobabamos los scores. Eso se hace a mano por el ejemplo, pero GridSearchCV lo hace por nosotros. 

Entonces, GridSearchCV toma: 
1. El modelo 
2. una rejilla de hiperparámetros
3. Una estrategia de cross-validation 
Entonces, prueba todas las combinaciones de hiperparametros (potente)
para cada combinación aplica crossvalidation 
calcula el score promedio
Luego escoje la mejor combinación 
y luego entrega un modelo final con esa mejor combinación jajajaj. Básicamente una sola función de librería resuelve un montón de problemas. 
Un grid (como vimos ya en el ejemplo) se ve así: 

```python
param_grid = {
    "n_estimators": [100,200,300],
    "max_depth": [None, 5, 10],
    "min_samples_split":[2,4]
    }
```

Entonces, tenemos 3 opciones para n_estimators, 3 para max_depth, 2 para min_samples_split, entonces, tenemos 3x3x6 = 18 configuraciones principales, y si usamos cross-validation con k=5, entonces: 18x5 = 90 entrenamientos. 
Entonces. ahi nos damos cuenta de la potencia de esa función. Aquí un ejemplo: 

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV, train_test_split

# 1. Separar train y test
X_train, X_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=42 
)

# 2. Definir modelo base
model = RandomForestClassifier(random_state=42)

# 3. Definir grid de hiperparámetros
param_grid = {
    "n_estimators": [100, 200, 300],
    "max_depth": [None, 5, 10],
    "min_samples_split": [2, 4, 6]
}

# 4. Grid search con cross-validation
grid = GridSearchCV(
    estimator=model,
    param_grid=param_grid,
    cv=5,
    scoring="accuracy", #La metrica que queremos maximizar
    n_jobs=-1 #Usamos todos los núcleos del procesador
)

# 5. Ajustar sobre train
grid.fit(X_train, y_train)

# 6. Mejor combinación encontrada
print("Best params:", grid.best_params_)
print("Best CV score:", grid.best_score_)

# 7. Evaluación final en test
best_model = grid.best_estimator_
test_score = best_model.score(X_test, y_test)
print("Test score:", test_score)
```
