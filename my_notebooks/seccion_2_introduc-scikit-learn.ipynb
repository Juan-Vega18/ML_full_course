Ahora, algo que es absolutamente importante en el ML es entender que la data debe ser divida en dos grupos, training set and test set. Es decir, un grupo de datos se usará para entrenar el modelo, el otro para probarlo. Usualmente se combina en proporciones como 80/20, 70/30, 75/25, tambien es posible tener el validation set que es un dataset extra que se usa durante el entrenamiento pero por ahora no lo vamos a usar. 

Scikit-learn ya incluye sklearn.model_selection.train_test_split que es la funcion de la libreria para dividir el dataset en training sets.

En el caso de este proyecto, estamos trabajando con un clasificador, booleano. Si o no. Asi que el curso nos recomienda usar empezar con el classifier Random forest.

```python
from sklearn.ensemble import RandomForestClassifier #Aqui hay muchos interesantes 

clf = RandomForestClassifier() #Aqui inicializamos el modelo, y vamos a ver los parametros
#que podemos modificar 
clf.get_params()
```

#Por ahora, 
no vamos a tocarlos, los dejaremos para mas adelante. Usaremos entonces
#los parametros por defecto de Sckit-learn. 

Ahora, vamos a Fit the model to the data and use it to make a prediction. Es lo que queremos. Si tenemos etiquetas (aprendizaje supervisado) el modelo trabaja en encontrar la relacion entre los datos y las etiquetas. En el caso de que no hayan etiquetas, el modelo intenta encontrar patrones y agrupar ejemplos similares juntos. 
La mayoria de los modelos de Scikit-Learn tienen el methodo fit(X,Y) integrado, donde X son las caracteristicas (features) e y son las etiquetas. 
En nuestro caso, vamos a empezar ajustando un modelo a los datos de entrenamiento separados. 



```python

clf.fit(X=X_train, y=y_train) #clf es nuestro modelo, y a ese modelo aplicamos parametros
```

Ahora, ya que tenemos el "modelo ajustado" viene la parte central de esto. intentar hacer una predicción. Despues de que el modelo ya está entrenado, ahora lo que hay que hacer es predecir, usando predict(). En otras palabras, usar el modelo con data nueva, que no haya visto y que no esté etiquetada. (Ojo, aqui es importente), la data que daremos para predecir debe tener el mismo shape que la data que usamos para entrenar (matematicamente tiene todo el sentido pero igual se aclara). 

Para eso, usaremos la data de test

```python
y_preds = clf.predict(X=X_test)
print(y_preds)

#Y si queremos analizar una unica fila, dado que es un 
#dataset 2D esperamos esa entrada

predic_1 = clf.predict(X_test.iloc[[0]])
predic_1
```

Ahí tenemos el array de predicciones, ahora lo que nos interesa es evaluar el modelo. Cada modelo estimador tiene un metodo integrado llamada score(). Este metodo compara que tan bien le fue al modelo aprendiendo los patrones entre las caracteristicas y las etiquetas. El metodo score() usa una metrica estandar de evaluacion para medir los resultados del modelo. 

En el caso de un clasificador (nuestro modelo), una de las metricas de evaluacion mas comunes es 'accuracy', que es la fraccion de predicciones correctas sobre el total de predicciones. 

```python
#Evaluate the model on the training set

train_acc = clf.score(X=X_test, y=y_test)
print(f'The models accuracy on the training dataset is: {train_acc*100} %')
```

Existen algunos otros metodos de evaluacion para nuestro modelo, estos vienen del modulo sklearn.metrics y tenemos: 
Classification report (y true, y true)
Construye un reporte mostrando varias metricas de clasificacion, como precision, recall, and F1-score. (luego las entenderemos)

Confusion matrix(y true, y pred). Crea una matriz de confusion para comparar la predición de las etiquetas verdaderas

Accuracy score (y_true, y_pred). Encuentra el puntaje de precisión, esta es la metrica por defecto de los clasificadores. 

```python
#Classification report 
from sklearn.metrics import classification_report

print(classification_report(y_test, y_preds))
```

```python
#Confusion matrix
from sklearn.metrics import confusion_matrix

print(confusion_matrix(y_test, y_preds))
```

```python
#Compute the accuracy score

from sklearn.metrics import accuracy_score

accuracy_score(y_test, y_preds)
```
