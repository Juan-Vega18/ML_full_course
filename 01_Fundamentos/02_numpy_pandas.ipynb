# 📊 NumPy y Pandas — Herramientas Esenciales para ML

Antes de entrenar modelos necesitamos dominar las herramientas de manipulación de datos:
- **NumPy**: operaciones numéricas y vectorizadas
- **Pandas**: DataFrames para análisis de datos tabulares

## 1. NumPy

NumPy introduce el objeto `ndarray`, un array N-dimensional eficiente que es la base de casi toda la computación científica en Python.

```python
import numpy as np

# Crear arrays
a = np.array([1, 2, 3, 4, 5])
b = np.arange(0, 10, 2)          # 0, 2, 4, 6, 8
c = np.linspace(0, 1, 5)         # 5 valores entre 0 y 1
zeros = np.zeros((3, 3))
ones  = np.ones((2, 4))
rand  = np.random.rand(3, 3)

print("a:", a)
print("b:", b)
print("c:", c)
print("zeros:\n", zeros)
print("rand:\n", rand.round(2))
```

```python
# Operaciones vectorizadas (sin bucles)
x = np.array([1, 2, 3, 4])

print("x + 10:   ", x + 10)
print("x * 2:    ", x * 2)
print("x ** 2:   ", x ** 2)
print("sqrt(x):  ", np.sqrt(x))
print("sum:      ", x.sum())
print("mean:     ", x.mean())
print("std:      ", x.std())
```

```python
# Álgebra lineal (clave en ML)
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])

print("A @ B (producto matricial):\n", A @ B)
print("Transpuesta de A:\n", A.T)
print("Determinante de A:", np.linalg.det(A))
print("Inversa de A:\n", np.linalg.inv(A).round(2))
```

```python
# Indexing y slicing
m = np.arange(1, 10).reshape(3, 3)
print("Matriz:\n", m)
print("Fila 0:", m[0])
print("Columna 1:", m[:, 1])
print("Submatriz:\n", m[:2, 1:])
print("Máscara booleana (>5):", m[m > 5])
```

## 2. Pandas

Pandas introduce `Series` (1D) y `DataFrame` (2D), ideales para datos tabulares como los que usaremos en proyectos de ML.

```python
import pandas as pd

# Crear DataFrame desde diccionario
datos = {
    'nombre': ['Ana', 'Luis', 'María', 'Carlos', 'Elena'],
    'edad':   [23, 35, 28, 45, 31],
    'salario':[30000, 55000, 42000, 70000, 48000],
    'ciudad': ['Madrid', 'Barcelona', 'Madrid', 'Valencia', 'Sevilla']
}
df = pd.DataFrame(datos)
df
```

```python
# Exploración básica
print("Shape:", df.shape)
print("\nInfo:")
df.info()
print("\nEstadísticas:")
df.describe()
```

```python
# Selección y filtrado
print("Columna edad:\n", df['edad'].values)
print("\nMayores de 30:")
print(df[df['edad'] > 30])
print("\nDe Madrid:")
print(df[df['ciudad'] == 'Madrid'])
```

```python
# Manipulación: nueva columna y agrupación
df['salario_mensual'] = df['salario'] / 12

print("Salario promedio por ciudad:")
print(df.groupby('ciudad')['salario'].mean().sort_values(ascending=False))
```

```python
# Manejo de valores nulos (muy común en ML)
df_con_nulos = df.copy()
df_con_nulos.loc[1, 'edad'] = None
df_con_nulos.loc[3, 'salario'] = None

print("Nulos por columna:")
print(df_con_nulos.isnull().sum())

# Estrategias: rellenar con la media o eliminar filas
df_limpio = df_con_nulos.copy()
df_limpio['edad'].fillna(df_limpio['edad'].mean(), inplace=True)
df_limpio.dropna(inplace=True)

print("\nDataFrame limpio:")
df_limpio
```

## 3. Resumen

| Herramienta | Para qué sirve en ML |
|-------------|----------------------|
| NumPy | Operaciones matriciales, features numéricas, álgebra lineal |
| Pandas | Cargar datos, EDA, limpieza, preprocesamiento |
