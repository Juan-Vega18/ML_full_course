# 📈 Visualización de Datos — Matplotlib y Seaborn

La visualización es fundamental en el Análisis Exploratorio de Datos (EDA). Permite entender distribuciones, detectar outliers y relaciones entre variables antes de modelar.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris

# Estilo global
sns.set_theme(style='whitegrid')
plt.rcParams['figure.figsize'] = (10, 5)

# Dataset
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['especie'] = [iris.target_names[i] for i in iris.target]
df.head()
```

## 1. Gráficos Básicos con Matplotlib

```python
# Histograma — distribución de una variable
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

axes[0].hist(df['sepal length (cm)'], bins=20, color='steelblue', edgecolor='white')
axes[0].set(title='Distribución: Longitud del Sépalo', xlabel='cm', ylabel='Frecuencia')

axes[1].hist(df['petal length (cm)'], bins=20, color='coral', edgecolor='white')
axes[1].set(title='Distribución: Longitud del Pétalo', xlabel='cm', ylabel='Frecuencia')

plt.tight_layout()
plt.show()
```

```python
# Scatter plot — relación entre dos variables
colores = {'setosa': 'blue', 'versicolor': 'green', 'virginica': 'red'}

plt.figure(figsize=(8, 5))
for especie, grupo in df.groupby('especie'):
    plt.scatter(
        grupo['petal length (cm)'], grupo['petal width (cm)'],
        label=especie, color=colores[especie], alpha=0.7
    )

plt.xlabel('Longitud del pétalo (cm)')
plt.ylabel('Ancho del pétalo (cm)')
plt.title('Relación: Longitud vs Ancho del Pétalo por Especie')
plt.legend()
plt.show()
```

## 2. Gráficos Estadísticos con Seaborn

```python
# Boxplot — distribución por categorías y detección de outliers
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

sns.boxplot(data=df, x='especie', y='sepal length (cm)', palette='Set2', ax=axes[0])
axes[0].set_title('Longitud del Sépalo por Especie')

sns.boxplot(data=df, x='especie', y='petal length (cm)', palette='Set2', ax=axes[1])
axes[1].set_title('Longitud del Pétalo por Especie')

plt.tight_layout()
plt.show()
```

```python
# Heatmap de correlaciones — clave para feature selection
plt.figure(figsize=(8, 6))
corr = df.drop(columns='especie').corr()
sns.heatmap(corr, annot=True, fmt='.2f', cmap='coolwarm', linewidths=0.5)
plt.title('Matriz de Correlaciones — Iris Dataset')
plt.show()
```

```python
# Pairplot — todas las relaciones de una vez
sns.pairplot(df, hue='especie', palette='Set1', diag_kind='kde', plot_kws={'alpha': 0.6})
plt.suptitle('Pairplot — Iris Dataset', y=1.02)
plt.show()
```

## 3. Conclusiones del EDA

A partir de la visualización podemos concluir:

1. **Setosa** es claramente separable de las otras dos especies por el tamaño del pétalo.
2. **Versicolor** y **Virginica** tienen cierto solapamiento → la clasificación entre ellas es más difícil.
3. Las variables de **pétalo** (longitud y ancho) tienen mayor poder discriminativo que las de sépalo.
4. Existe alta **correlación** entre longitud y ancho del pétalo → podríamos considerar reducción de dimensionalidad.
